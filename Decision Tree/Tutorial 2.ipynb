```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(data = iris.data , columns = iris.feature_names)
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split

x = iris.data
y = iris.target

#splitting the data to train test split 

x_train , x_test , y_train , y_test = train_test_split( x , y , test_size = 0.25 , random_state = 42)
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train , y_train)
y_pred = dt.predict(x_test)
```

```python
#checking the accuracy of it 

from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pred)
```

NOW WE WILL BE CREATING A VISUAL REPRESENTATION OF A DECISION TREE MODEL . THE RESULTING PLOT SHOWS THE DECISIONS MADE AT EACH NODE , THE FEATURES USED TO SPLIT 
THE DATA & R THE DISTRIBUTION OF CLASSES OR VALUES AT EACH NODE .

Visual representation of a decision tree model: The script generates a plot (or a graphical representation) of a decision tree. Each internal node of the tree corresponds to a decision based on one of the input features, and the branches from that node correspond to the possible values of that feature. The leaf nodes (the terminal nodes of the tree) represent the output values (or classes) of the model.

Decisions made at each node: In the plot, each internal node represents a decision point where the data is split. The decision is based on the value of one of the input features. For example, a node might represent a decision like “Is feature X > 1.5?” with two branches coming off the node representing “yes” and “no”.

Features used to split the data: The feature used to split the data at each node is chosen based on the algorithm used to train the decision tree. This is typically the feature that results in the largest information gain or the largest reduction in impurity. In the plot, the feature used to split the data at each node is typically indicated at the top of the node.

Distribution of classes or values at each node: In a classification problem, each leaf node in the plot represents a class, and the number of samples of each class that end up in that leaf node can be represented by the color of the node. In a regression problem, each leaf node represents a value, and the range of values that end up in that leaf node can be represented by the color of the node.

So, in essence, the script is creating a visual way to understand how the decision tree model is making its predictions, by showing the decisions made at each node, the features used to make those decisions, and the resulting distribution of classes or values.

```python
from sklearn import tree

#checkingthe split with gini impurity test 
dt1 = DecisionTreeClassifier(criterion = 'gini')
dt1.fit(iris.data , iris.target)
plt.figure(figsize= (15,10))
tree.plot_tree(dt1, filled = True)
plt.show()
```

```python
#checking the split with entropy impurity

dt2 = DecisionTreeClassifier(criterion = 'entropy')
dt2.fit(iris.data , iris.target)
plt.figure(figsize= (15,10))
tree.plot_tree(dt2, filled = True)
plt.show()
```
